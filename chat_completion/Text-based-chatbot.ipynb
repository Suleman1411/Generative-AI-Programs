```python
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
import os

_: bool = load_dotenv(find_dotenv())
client: OpenAI = OpenAI()
```

```python
from openai.types.chat.chat_completion import ChatCompletion
# You have the flexibility to customize the token size according to your 
# requirements byadjusting the 'max_tokens' parameter, which is currently set to 300.
# Just modify int = 300 to any number that you want

def chat_completion(prompt: str, max_tokens: int = 300) -> str:
    response: ChatCompletion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        model="gpt-3.5-turbo-1106",
        max_tokens=max_tokens,
    )
    return response.choices[0].message.content

def main():
    use_cases_list = [
        "Code generation",
        "Language Translation",
        "Storytelling",
        "Interview practice",
        "Recipe Ideas"
    ]

    print("Choose a use case:")
    for i, use_case in enumerate(use_cases_list, start=1):
        print(f"{i}. {use_case}")

    user_choice = int(input("Enter the number of the use case you want to try: "))

    if 1 <= user_choice <= 5:
        selected_use_case = use_cases_list[user_choice - 1]
        print(f"You've selected: {selected_use_case}")
        prompt_choice = input("Would you like to provide additional details for improved chatbot performance? (yes/no): ").lower()

        if user_choice == 1:
            if prompt_choice == "yes" or prompt_choice == "y":
                target_programming_lan = input("Enter the target programming language: ")
                program_name = input("Specify the program you want to code: ")
                additional_choice = input("Do you want to add more context? (yes/no): ").lower()
                context = f" that {input('Specify more details: ') if additional_choice == 'yes' or additional_choice == 'y' else ''}"
                prompt = f"Generate a {target_programming_lan} code snippet for a {program_name} {context}"
            else:
                prompt = "Assist me in generating a code snippet:"

        elif user_choice == 2:
            if prompt_choice == "yes" or prompt_choice == "y":
                target_language = input("Enter your target language: ")
                sentence_choice = input("Do you have a sentence to translate? (yes/no): ").lower()
                sentence = input("Enter the sentence in English: ") if sentence_choice == "yes" or sentence_choice == "y" else ''
                prompt = f"Translate the following sentence into {target_language}: {sentence}"
            else:
                prompt = "Assist me in translating a sentence. Provide the sentence in English, and specify the target language for translation."

        elif user_choice == 3:
            if prompt_choice == "yes" or prompt_choice == "y":
                target_story = input("Enter the name of the story: ")
                plot_choice = input("Do you want to provide additional plot details? (yes/no): ").lower()
                plot_details = f" where {input('Enter plot details: ') if plot_choice == 'yes' or plot_choice == 'y' else ''}"
                prompt = f"Create a captivating story involving {target_story}{plot_details}."
            else:
                prompt = "Assist me in writing a story."

        elif user_choice == 4:
            if prompt_choice == "yes" or prompt_choice == "y":
                target_interview = input("Enter the target interview position: ")
                topics_choice = input("Do you have specific interview topics? (yes/no): ").lower()
                interview_topics = f" related to {input('Enter interview topics: ') if topics_choice == 'yes' or topics_choice == 'y' else ''}"
                prompt = f"Conduct a simulated interview for a {target_interview} position {interview_topics}."
            else:
                prompt = "Simulate an interview."

        elif user_choice == 5:
            if prompt_choice == "yes" or prompt_choice == "y":
                target_dish = input("Enter the target dish: ")
                ingredients_choice = input("Do you have a list of ingredients? (yes/no): ").lower()
                ingredients_list = f" that includes {input('Enter ingredients: ') if ingredients_choice == 'yes' or ingredients_choice == 'y' else ''}"
                prompt = f"Suggest a unique and flavorful recipe for {target_dish}{ingredients_list}."
            else:
                prompt = "Help me a recipe idea."

        user_input = input(f"{prompt} ")
        response = chat_completion(user_input)
        print("Chatbot Response:", response)
    else:
        print("Invalid choice. Please choose a number between 1 and 5.")

if __name__ == "__main__":
    main()
```
